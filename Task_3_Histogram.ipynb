```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ds = pd.read_csv("Iris.csv")
ds
```

```python
ds.shape
```

```python
ds.info()
```

```python
ds.describe()
```

```python
print(f'Number of missing values = {ds.isna().sum().sum()}')
print(f'Number of duplicates = {ds.duplicated().sum()}')
```

```python
ds.drop_duplicates(inplace=True)
```

```python
ds
```

```python
sns.pairplot(ds,hue='Species',size=4)
plt.show()
```

```python
sns.violinplot(x='Species',y='PetalLengthCm',data=ds)
plt.show()
```

```python
sns.violinplot(x='Species',y='SepalLengthCm',data=ds)
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(2,2,1)
sns.histplot(ds['PetalLengthCm'],bins=20,ax=ax1)
ax1.set_xlabel('Petal Length (Cm)')
ax1.set_title('Distribution Of Petal Length')
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
ax2 = fig.add_subplot(2,2,2)
sns.histplot(ds['PetalWidthCm'],bins=20,ax=ax2)
ax2.set_xlabel('Petal Width (Cm)')
ax2.set_title('Distribution Of Petal Width')
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
ax3 = fig.add_subplot(2,2,3)
sns.histplot(ds['SepalLengthCm'],bins=20,ax=ax3)
ax3.set_xlabel('Sepal Length (Cm)')
ax3.set_title('Distribution Of Sepal Length')
plt.show()
```

```python
fig = plt.figure(figsize=(10,7))
ax4 = fig.add_subplot(2,2,4)
sns.histplot(ds['SepalWidthCm'],bins=20,ax=ax4)
ax4.set_xlabel('Sepal Width (Cm)')
ax4.set_title('Distribution Of Sepal Width')
plt.tight_layout()
plt.show()
```
